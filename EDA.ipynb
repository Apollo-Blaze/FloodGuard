```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

Loading the dataset

```python
file_path = 'E:\\project\\FloodGuard\\final.csv'  
df = pd.read_csv(file_path)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='disaster',data = df)
plt.title('Distribution of disasters')
plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['severity'])
plt.title('Distribution of Severity')
plt.show()
#severity is 0 -> no severe
#severity is 0.5 -> disaster \ disaster info
#severity is 1 ->both severity
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['tavg'],kde = True)
plt.title('Average Temperature Distribution')
plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='disaster', y='severity', data=df)
plt.title('Severity of Different Disaster Types')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tavg', y='severity', hue='disaster', data=df)
plt.title('Average Temperature vs. Severity')
plt.show()
```

 Multivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tavg', y='prcp', size='severity', hue='disaster', data=df)
plt.title('Temperature, Precipitation, and Severity of Disasters')
plt.show()
```

Correlation Matrix

```python
corr = df[['tavg', 'tmin', 'tmax', 'prcp', 'snow', 'wdir', 'wspd', 'pres', 'severity']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix ')
plt.show()
```

```python
correlation_with_severity = corr['severity'].sort_values(ascending=False)
print(correlation_with_severity)
```

Feature Selection


Based on the correlation matrix, we  select highly correlated features with severity for modeling like 'tavg', 'prcp', 'wspd'
